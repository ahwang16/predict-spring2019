```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
with open('./datasets/idiomLexiconScored.json', 'r') as infile:
    lex = json.load(infile)
```

```python
plt.rc('axes', titlesize=16)
plt.rc('axes', labelsize=16)
```

```python
pos, pos_label = [], []
neg, neg_label = [], []
neu, neu_label = [], []
for idiom in lex:
    if lex[idiom][2] == 'inappropriate':
        continue
    elif lex[idiom][2] == 'positive':
        if lex[idiom][0][0] is None:
            continue
        pos.append(lex[idiom][0][0])
        pos_label.append(lex[idiom][1])
    elif lex[idiom][2] == 'negative':
        if lex[idiom][0][1] is None:
            continue
        neg.append(lex[idiom][0][1])
        neg_label.append(lex[idiom][1])
    else:
        if lex[idiom][0][2] is None:
            continue
        neu.append(lex[idiom][0][2])
        neu_label.append(lex[idiom][1])
        
        
```

```python
np.mean(neu)
```

```python
np.std(neu)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# ax.scatter(pos_label, pos, marker='o', label='positive')
# ax.scatter(neg_label, neg, marker='s', label='negative', c='orange')
ax.scatter(neu_label, neu, marker='x', label='neutral', c='green')
plt.title('Sentiment Scores for Neutrally Labeled Idioms')
plt.xlabel('SLIDE Positive Percent')
plt.ylabel('DAL Positive Index')
# plt.legend()
# plt.subplots_adjust(left=0, bottom=0, right=1, top=1.3)
plt.show()
# plt.savefig('neutral.png', bbox_inches = "tight")
```
